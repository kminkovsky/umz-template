```python
%matplotlib inline

import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

train = pd.read_csv('train/in.tsv', sep='\t', names=['price','mileage','year','brand','engingeType','engineCapacity'])
sns.lmplot(x='engineCapacity', y='price', data=train)
```

```python
lr = LinearRegression()
lr.fit(train[['mileage', 'year', 'engineCapacity']].values.reshape(-1, 3), train['price'].values.reshape(-1, 1))
dev = pd.read_csv('dev-0/in.tsv', sep='\t', names=['mileage','year','brand','engingeType','engineCapacity'])
out = pd.DataFrame(lr.predict(dev[['mileage', 'year', 'engineCapacity']].values.reshape(-1, 3)))
out.to_csv('dev-0/out.tsv', sep='\t', header=False, index=False)
```
