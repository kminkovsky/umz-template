```python
%matplotlib inline

import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

train = pd.read_csv('train/train.tsv', sep='\t', names=['price', 'isNew', 'rooms', 'floor', 'location', 'sqrMetres'])
sns.lmplot(x='sqrMetres', y='price', data=train)
```

```python
lr = LinearRegression()
lr.fit(train[['sqrMetres', 'floor']].values.reshape(-1, 2), train['price'].values.reshape(-1, 1))
dev = pd.read_csv('dev-0/in.tsv', sep='\t', names=['isNew', 'rooms', 'floor', 'location', 'sqrMetres'])
out = pd.DataFrame(lr.predict(dev[['sqrMetres', 'floor']].values.reshape(-1, 2)))
out.to_csv('dev-0/out.tsv', sep='\t', header=False, index=False)
```
